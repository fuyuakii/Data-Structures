> #### **2.2.1 Sobre as distâncias**

Foi utilizado o método de Haversine para calcular as distâncias entre todos os pares de vértices (focos-focos e bases-focos). A fórmula de Haversine considera a curvatura da Terra:

$$
d = 2R \arcsin\!\left(
    \sqrt{
        \sin^2\!\left(\frac{\Delta\varphi}{2}\right)
        +
        \cos(\varphi_1)\cos(\varphi_2)\sin^2\!\left(\frac{\Delta\lambda}{2}\right)
    }
\right)
$$

> #### **3.3 Implementação do Grafo**

A implementação usa uma lista de adjacência na classe Vertice, escolhida por ser eficiente para grafos esparsos (muitos vértices). A classe abaixo foi estuda em sala de aula, apenas com algumas modificações.

```python
class Vertice:
    def __init__(self, nome):
        self.nome = nome
        self.listaArestas = []
        self.custo = float('inf')
        self.estimativa = 0
        self.prioridade = float('inf')
        self.anterior = None

    def adicionarArestaUnica(self, outro, peso):
        if not any(v == outro for v, _ in self.listaArestas):
            self.listaArestas.append((outro, peso))

    def adicionarArestaBidirecionalUnica(self, outro, peso):
        self.adicionarArestaUnica(outro, peso)
        outro.adicionarArestaUnica(self, peso)

    def apagarAresta(self, vertice):
        novaLista = [(v, d) for (v, d) in self.listaArestas if v != vertice]
        if len(novaLista) != len(self.listaArestas):
            self.listaArestas = novaLista
            vertice.listaArestas = [
                (v, d) for (v, d) in vertice.listaArestas if v != self
            ]

    def __str__(self):
        return self.nome
```

> #### **3.4 Cálculo das Distâncias**

Esta etapa utiliza a fórmula de Haversine para calcular as distâncias geográficas entre os pontos.

Também são calculadas as conexões seguindo as regras estabelecidas: cada brigada conecta-se ao foco mais próximo, e cada foco conecta-se aos três focos mais próximos.


```python
def calcularHaversine(lat1, lon1, lat2, lon2):
    raioTerra = 6371
    lat1Rad = math.radians(lat1)
    lon1Rad = math.radians(lon1)
    lat2Rad = math.radians(lat2)
    lon2Rad = math.radians(lon2)
    deltaLat = lat2Rad - lat1Rad
    deltaLon = lon2Rad - lon1Rad
    a = (math.sin(deltaLat / 2) ** 2 +
         math.cos(lat1Rad) * math.cos(lat2Rad) * math.sin(deltaLon / 2) ** 2)
    c = 2 * math.asin(math.sqrt(a))
    return raioTerra * c

def obterCoordenadas(verticeObj, focosCoords, brigadasCoords):
    if verticeObj.nome.startswith('V'):
        indice = int(verticeObj.nome[1:])
        return focosCoords[indice]

    elif verticeObj.nome.startswith('B'):
        indice = int(verticeObj.nome[1:])
        return brigadasCoords[indice]

    return None

def calcularHeuristica(verticeAtual, verticeDestino):
    coordsAtual = obterCoordenadas(verticeAtual, focos, brigadas)
    coordsDestino = obterCoordenadas(verticeDestino, focos, brigadas)

    if coordsAtual and coordsDestino:
        return calcularHaversine(coordsAtual[0], coordsAtual[1],
                                 coordsDestino[0], coordsDestino[1])
    return 0


brigadasParaFocos = []

for i, brigada in enumerate(brigadas):
    distancias = []

    for j, foco in enumerate(focos):
        distancia = calcularHaversine(brigada[0], brigada[1], foco[0], foco[1])
        distancias.append((f"V{j}", distancia))

    distancias.sort(key=lambda x: x[1])
    brigadasParaFocos.append((f"B{i}", distancias[0][0], distancias[0][1]))


focosParaFocos = []

for i in range(len(focos)):
    distancias = []

    for j in range(len(focos)):
        if i != j:
            distancia = calcularHaversine(focos[i][0], focos[i][1],
                                          focos[j][0], focos[j][1])
            distancias.append((f"V{j}", distancia))

    distancias.sort(key=lambda x: x[1])
    focosParaFocos.append((f"V{i}", [distancias[0], distancias[1], distancias[2]]))
```

> ### **3.5 Adiciona os Vértices e Arestas**

Nesta etapa, o grafo ganha forma. No código, é criado os objetos `Vertice` para todos os focos e brigadas e guardamos tudo em um dicionário para operações de acesso. Depois, usando as listas de conexões já calculadas (`brigadasParaFocos` e `focosParaFocos`), adicionamos as arestas: cada brigada liga-se ao foco mais próximo e cada foco liga-se aos três focos mais próximos.

```python
vertices = {}

for i in range(len(focos)):
    vertices[f"V{i}"] = Vertice(f"V{i}")

for i in range(len(brigadas)):
    vertices[f"B{i}"] = Vertice(f"B{i}")

for brigada, foco, distancia in brigadasParaFocos:
    vertices[brigada].adicionarArestaUnica(vertices[foco], distancia)

for focoOrigem, conexoes in focosParaFocos:
    for focoDestino, distancia in conexoes:
        vertices[focoOrigem].adicionarArestaBidirecionalUnica(vertices[focoDestino], distancia)
```

# **5. Aplicação de Algoritmo de Busca**

-----


> ## **5.1 Algoritmo `A*`**

`A*` é um algoritmo de busca de caminhos que tem como objetivo principal encontrar a rota mais curta entre dois pontos em um determinado grafo. Ele busca encontrar o menor custo para ir do nó presente até o nó objetivo. Um dos seus pilares é explorar o mínimo de nós possíveis.

Ele é uma extensão do algoritmo de Dijkstra, pois combina esse algoritmo com uma propriedade chamada heurística. A heurística é uma função que busca estimar o custo restante para ir do nó presente até seu ponto final. A grande vantagem do uso da heurística é que ela explora menos caminhos do que o Dijkstra, tornando mais rápido a busca.

Existem vários tipos de heurística voltadas a distâncias, como Manhattan, Euclidiana e Chebyshev. A escolha certa depende do problema. A função que enraíza esse algoritmo é:

**f(n) = g(n) + h(n)**

Onde:

- **g(n)**: custo real para chegar até o nó de destino;
- **h(n)**: custo estimado (heurística) do nó presente até o nó de destino.

O algoritmo sempre busca expandir o próximo nó com menor valor de **f(n)**.

> ## **5.2 Implementação do `A*`**

- **limparVertices()**: reinicia os atributos dos vértices para uma nova busca.

- **construirCaminho()**: reconstrói o caminho ótimo a partir do vértice destino.

- **aEstrela()**: executa o algoritmo propriamente dito, utilizando uma fila de prioridades (heap) para expandir os vértices com menor custo estimado  
  \( f(n) = g(n) + h(n) \), onde \( h(n) \) é calculada pela função `calcularHeuristica`.


```python
def limparVertices(vertices):
    for vertice in vertices.values():
        vertice.custo = float('inf')
        vertice.anterior = None
        vertice.estimativa = 0
        vertice.prioridade = float('inf')

def construirCaminho(vertice):
    caminho = []
    atual = vertice
    while atual is not None:
        caminho.append(atual.nome)
        atual = atual.anterior
    return caminho[::-1]

def aEstrela(inicio, fim, vertices):
    limparVertices(vertices)

    inicio.custo = 0
    inicio.estimativa = calcularHeuristica(inicio, fim)
    inicio.prioridade = inicio.custo + inicio.estimativa

    abertos = []
    contador = 0
    heapq.heappush(abertos, (inicio.prioridade, contador, inicio))

    fechados = set()

    while abertos:
        _, _, atual = heapq.heappop(abertos)

        if atual in fechados:
            continue
        fechados.add(atual)

        if atual == fim:
            return construirCaminho(atual), atual.custo

        for vizinho, peso in atual.listaArestas:
            if vizinho in fechados:
                continue

            novoCusto = atual.custo + peso

            if novoCusto < vizinho.custo:
                vizinho.anterior = atual
                vizinho.custo = novoCusto
                vizinho.estimativa = calcularHeuristica(vizinho, fim)
                vizinho.prioridade = vizinho.custo + vizinho.estimativa

                contador += 1
                heapq.heappush(abertos, (vizinho.prioridade, contador, vizinho))

    return None, float('inf')
```

# **6. Análise do Grafo**

-----



## **Pergunta 1 — Qual a importância de modelar o cenário de queimadas como um grafo?**

> Ao modelar o problema como um grafo, cada ponto relevante do território (áreas de queimada, estradas, bases de brigada, cruzamentos, regiões críticas) torna-se um vértice, enquanto as possíveis rotas entre eles são arestas com pesos representando distância real. Isso permite que encontrem trajetos mais rápidos e eficientes. Sem a estrutura de grafo, seria muito mais difícil representar a conectividade entre as regiões e calcular rotas otimizadas de resposta.


## **Pergunta 2 — Por que o algoritmo `A*` foi escolhido em vez de outras alternativas?**

> O `A*` combina o custo real percorrido (g-score) com uma heurística admissível, garantindo rapidez e precisão ao buscar rotas em mapas espaciais. Diferente do Dijkstra, que explora muito mais caminhos, o `A*` prioriza direções promissoras usando a distância Euclidiana como heurística. No contexto de combate a incêndios, onde tempo de resposta é crítico, o `A*` oferece um equilíbrio ideal entre velocidade e qualidade da rota encontrada.

## **Pergunta 3 — Como a distribuição das brigadas influencia os caminhos encontrados?**

> As brigadas foram posicionadas próximas a regiões com maior concentração de vértices, o que reduz significativamente o custo total das rotas quando ocorre um foco de incêndio. Quando o grafo é analisado, percebe-se que vértices de brigada centrais (como B1 ou B2, dependendo do mapa) normalmente geram rotas mais curtas para vários pontos críticos. Essa distribuição estratégica diminui o tempo de deslocamento e melhora a eficiência da contenção.

## **Pergunta 4 — O grafo apresenta regiões críticas? Como identificá-las?**

> Sim. Regiões críticas podem ser identificadas observando vértices que são muito conectados (alto grau) ou vértices que funcionam como “pontes” entre regiões do grafo. Se um destes vértices falha (por barreira, estrada interditada, avanço do fogo), o acesso a grandes áreas fica comprometido. No grafo do notebook, alguns vértices de estrada intermediários têm papel estrutural importante, sendo essenciais para manter a conectividade entre bases de brigada e focos de incêndio.

## **Pergunta 5 — O caminho gerado pelo `A*` é sempre o mais seguro para brigadistas?**

> Não necessariamente. O `A*` encontra o menor caminho em termos de custo/distancia, mas a segurança envolve outros fatores que não estão no peso do grafo, como direção dos ventos, velocidade do fogo, tipo de relevo ou presença de obstáculos. Portanto, o `A*` fornece uma rota teoricamente mais rápida, mas ainda seria necessário incorporar ao grafo pesos adicionais (como risco, acessibilidade, condições ambientais) para que as rotas otimizadas também representem a escolha mais segura.

```python
import math
import pandas as pd

def get_coords_universal(item):
    nome = ""
    if hasattr(item, 'nome'): nome = item.nome
    elif isinstance(item, str): nome = item
    else: return None
    try:
        if nome.startswith('V'):
            idx = int(nome[1:])
            if idx < len(focos): return focos[idx]
        elif nome.startswith('B'):
            idx = int(nome[1:])
            if idx < len(brigadas): return brigadas[idx]
    except:
        return None
    return None

def calc_distancia_km(coord1, coord2):
    if not coord1 or not coord2: return 0
    R = 6371
    lat1, lon1 = math.radians(coord1[0]), math.radians(coord1[1])
    lat2, lon2 = math.radians(coord2[0]), math.radians(coord2[1])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c
```

## **Pergunta 6 — Qual é a distância total acumulada da rota otimizada?**

```python
print(">>> CÁLCULO DE CUSTO TOTAL DA ROTA")
if 'caminho' not in globals() or caminho is None:
    print("ERRO: Não encontrei a rota calculada.")
    print("DICA: Rode novamente a Célula do Algoritmo A* (Seção 5) antes desta.")
else:
    distancia_total = 0
    if len(caminho) > 1:
        print(f"Iniciando trajeto com {len(caminho)} pontos...")
        for i in range(len(caminho) - 1):
            u = caminho[i]
            v = caminho[i+1]

            c1 = get_coords_universal(u)
            c2 = get_coords_universal(v)

            d = calc_distancia_km(c1, c2)
            distancia_total += d

        print(f"------------------------------------------------")
        print(f"RESPOSTA: A distância total da missão é de {distancia_total:.2f} km.")
        print(f"------------------------------------------------")
    else:
        print("A rota tem menos de 2 pontos, distância é 0.")
```

## **Pergunta 7 — Qual é o foco de incêndio mais isolado (crítico)?**

```python
print(">>> ANÁLISE DE RISCO (FOCO MAIS ISOLADO)")

maior_distancia = -1
foco_critico = ""

if 'focos' in globals() and 'brigadas' in globals():
    for i, f_coord in enumerate(focos):
        min_dist_foco = float('inf')
        for b_coord in brigadas:
            d = calc_distancia_km(f_coord, b_coord)
            if d < min_dist_foco:
                min_dist_foco = d
        if min_dist_foco > maior_distancia:
            maior_distancia = min_dist_foco
            foco_critico = f"V{i}"

    print(f"RESPOSTA: O foco mais crítico é o {foco_critico}.")
    print(f"Ele está a {maior_distancia:.2f} km da base de ajuda mais próxima.")
    print("Recomendação: Criar nova base nesta região.")
else:
    print("Erro: Listas de Focos/Brigadas não encontradas.")
```

## **Pergunta 8 — Como gerar um relatório tabular (DataFrame) das distâncias?**

```python
print(">>> RELATÓRIO DE COBERTURA DA BASE B0")

dataset = []
if 'brigadas' in globals() and 'focos' in globals():
    base_ref = brigadas[0]

    for i, f_coord in enumerate(focos):
        d = calc_distancia_km(base_ref, f_coord)
        status = "Alta Prioridade" if d < 50 else ("Média" if d < 100 else "Monitoramento")

        dataset.append({
            "Base Origem": "Brigada B0",
            "Destino": f"Foco V{i}",
            "Distância (km)": round(d, 2),
            "Status Logístico": status
        })
    df = pd.DataFrame(dataset).sort_values(by="Distância (km)")
    display(df.head(10))
else:
    print("Dados insuficientes para gerar tabela.")
```
